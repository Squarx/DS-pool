## Task 1

Напиши код, который будет решать квадратное уравнение вида
$ax^2 + bx + c = 0$. \
Для этого понадобится найти дискриминант. Его формула - $ D = b^2 − 4ac $

Для начала напиши формулу нахождения корней квадратного уравнения на языке LaTeX

$ x = \frac{{-b} ± {\sqrt{D}}} {2a}\\ \\$
$D = b^2 - 4ac$

Теперь, когда ты вспомнил как находить корни, приступай к написанию функции:
- Функция принимает на вход коэффициент a, b, c.
- Функция должна возвращать кортеж корней уравнения. 
- Если корней у уравнения нет, то функция должна вернуть None

```python
from math import sqrt

def linear_solver(b, c):
    if (b != 0):
      return (-c / b)
    elif (c == 0):
      return 'Quantity of solutions: inf'
    else:
      return 'Zero solutions'

def quadratic_equation(a, b, c):
    if (a == 0):
        return linear_solver(b, c)
    
    d = b * b - 4 * a * c

    if (d == float('inf')):
        return float('inf')
    elif (d < 0):
        return 'Zero solutions'
    
    x1 = (-b + sqrt(d)) / (2 * a)
    x2 = (-b - sqrt(d)) / (2 * a)

    if (x1 - x2 > 1e-6):
        return tuple([round(x1, 2), round(x2, 2)])
    else:
        return x1
    pass

print("2x^2 + 8x + 2 = 0 \n Корни уровнения:", quadratic_equation(-4,28,-49))
print("2x^2 + 8x + 2 = 0 \n Корни уровнения:", quadratic_equation(1,-1,-0))

print("2x^2 + 8x + 2 = 0 \n Корни уровнения:", quadratic_equation(3,-4,94))

```

Написание проверки на пaлиндром.

1. На вход функции _isPalindrome_ подается **число** (int). \
Функция возвращает **True**, если число является палидромом и **False** в противном случае. 

```python
def isPalindrome(x: int) -> bool:
    string_number = str(x)

    if (string_number == string_number[::-1]):
        return 1
    else:
        return 0
    pass
print(isPalindrome(121))
print(isPalindrome(1234554321))
print(isPalindrome(1))
print(isPalindrome(12))
print(isPalindrome(1001))
print(isPalindrome(1221))
```
